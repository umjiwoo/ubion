1. corona.csv 파일 로드
2. 쓸모없는 컬럼 삭제

```python
import pandas as pd
import numpy as np
```

```python
corona = pd.read_csv("./csv/corona.csv")
corona.head(2)
```

```python
# 필요없는 컬럼 날리기 -> 이미 실행해서 날아간 경우 에러 발생할 수 있음
corona = corona.drop("Unnamed: 0",axis=1)
```

```python
corona.head()
```

```python
corona.columns=['등록일시', '누적사망자', '누적확진자', '게시글번호', '기준일', '기준시간','수정일시', '누적의심자', '누적확진률']
```

```python
corona.head()
```

```python
# 데이터프레임의 정렬을 변경하는 함수
# 1. sort_values() -> 매개변수 종류 ascending=True(dafault) inplcae(기준이 되는 데이터프레임을 수정할 것인가의 여부)=False(default)
# inplcae 옵션 지정 시 해당 데이터프레임이 수정된 데이터프레임으로 set됨//
# inplcae 옵션 사용 시 corona = corona.sort_values("정렬 키",inplace=True)로 안하고 corona.sort_values("정렬 키",inplace=True)만 해줘도 ok
corona.sort_values("등록일시", inplace=True)
corona
```

```python
# 인덱스 초기화 -> reset_index()
# 매개변수 -> drop(기존 인덱스에 존재하던 값을 삭제할 것인가의 여부)=False(default)  inplcae
corona.reset_index(drop=True, inplace=True)
# 위아래 동일
# corona.reset_index().drop('index',axis=1, inplace=True)
```

```python
# 데이터 간의 차이를 출력하는 함수
# diff(n)
# n=1(default) ; 하나 전의 인덱스 데이터를 선택해 차를 구함

corona.누적확진자.diff(1)
```

```python
corona['일일확진자'] = corona.누적확진자.diff(1)
```

```python
# 이동한 값을 출력 -> shift()
corona2 = corona.head(5)
corona2
```

```python
corona['일일사망자'] = corona['누적사망자'] - corona['누적사망자'].shift(1)
```
