```python
import pandas as pd
```

```python
sales = pd.read_csv("./csv/Sales Records.csv")
sales
```

```python
sales.info()
```

```python
# sales.drop("Order ID", axis=1,inplace=True)
# sales.drop("Total Revenue", axis=1,inplace=True)
# sales.drop("Total Cost", axis=1,inplace=True)
# sales.drop("Total Profit", axis=1,inplace=True)
# sales.drop("Sales Channel", axis=1,inplace=True)
sales.drop(["Order ID","Total Revenue","Total Cost","Total Profit","Sales Channel"], axis=1, inplace=True)
```

```python
sales.columns=['권역', '국가', '상품종류', '우선순위', '주문일자','발송일자', '판매단위', '단가', '원가']
```

```python
sales.sort_values(["권역","국가"],inplace=True)
```

```python
sales.reset_index(drop=True, inplace=True)
```

```python
sales
```

```python
# 새로운 파생 변수 생성 
# 총단가 = 판매단위 x 단가
# 총원가 = 판매단위 x 원가
# 총이윤 = 총단가 - 총원가

sales["총단가"] = sales['판매단위'] * sales['단가']
sales["총원가"] = sales['판매단위'] * sales['원가']
sales["총이윤"] = sales['총단가'] - sales['총원가']

```

```python
sales
```

```python
sales.tail()
```

```python
# 데이터 분포 파악

sales["국가"].value_counts()
```

```python
sales["국가"].value_counts().index
```

```python
_list = sales["국가"].unique()
```

```python
sales.loc[sales['국가']=="South Korea"]['총이윤'].sum()
```

```python
profit_list = {}

for i in _list:
    total_profit = sales.loc[sales["국가"]==i]["총이윤"].sum()
    profit_list[i]=total_profit

profit_list
```

```python
pd.Series(profit_list).sort_values(ascending=False)
```

```python
sales["국가 총이윤"] = 0
```

```python
for i in _list:
    sales.loc[sales["국가"]==i,"국가 총이윤"] = profit_list[i]
```
