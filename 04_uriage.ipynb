```python
uriage.drop("item_name", axis=1, inplace=True)
```

```python
uriage = origin_data.copy()
uriage.info()
```

```python
uriage["item_name"] = uriage["item_name"].str.upper()
```

```python
uriage
```

```python
def change(x):
    result= x.lower()
    return result

# 데이터프레임의 열에 대해 함수를 적용하기 위해서는 apply() 이용해야
uriage["item_name"] = uriage["item_name"].apply(change)
```

```python
# 람다함수로 적용
uriage["item_name"] = uriage["item_name"].apply(lambda x:x.upper())
```

```python
uriage
```

```python
# item_name에 있는 공백을 제거 한다.
# 공백 제거 함수 : strip() -> 문자열 좌우 공백 제거
# 현재 데이터에서는 strip() 사용 불가
# replace(기준문자열, 대체할문자열)
# replace(" ", "")

def change(x):
    result = x.replace(" ","")
    result = result.lower()
    return result

uriage["item_name"]=uriage["item_name"].apply(change)

#uriage['item_name'] = uriage['item_name'].apply(lambda x : x.replace(" ", "").upper())

uriage.head()
```

```python
# 각 행이 item_price에 대한 결측치를 가지는 지에 대한 여부
flg_null = uriage["item_price"].isna()

# flg_null이 True를 가지는 행에 대해서만 추출
uriage.loc[flg_null]
```

```python
# item_name이 "상품A"인 경우 추출
uriage.loc[uriage["item_name"]=="상품a"]
```

```python
#  item_name이 "상품A"이고 item_price에 결측치를 갖는 경우 추출
uriage.loc[(flg_null) & (uriage["item_name"]=="상품a")]
```

```python
uriage
```

```python
uriage.loc[(uriage["item_name"]=="상품a") & ~(flg_null), "item_price"]
```

```python
uriage.loc[(uriage["item_name"]=="상품a") & ~(flg_null), "item_price"].value_counts()
```

```python
price = uriage.loc[(uriage["item_name"]=="상품a") & ~(flg_null), "item_price"].value_counts().index[0]
price
```

```python
# fillna(n) -> n값으로 Nan값 채움
uriage.loc[(flg_null)&(uriage["item_name"]=="상품a")].fillna(price)

#uriage.loc[(flg_null) & (uriage['item_name'] == '상품a'), 'item_price'] = price (위와 동일)
```
