```python
import requests
import xmltodict
from urllib import parse
import pandas as pd
```

```python
url = 'http://apis.data.go.kr/1160100/service/GetFnCoBasiInfoService/getFnCoOutl'
params ={
    'serviceKey' : '/VOjiRaIppLwgf1mXaAkdM6WBLZ/OoWeirK2V1U92jDVjSTAmj16PHQhpVYuumy3EeiuxjYGQxMleNu17eLAyA==',
    'pageNo' : '1', 
    'numOfRows' : '10',
    'resultType' : 'xml',
    'basDt' : '20200408',
    'crno' : '1101113892240',
    'fncoNm' : '메리츠자산운용' 
    }

response = requests.get(url, params=params)
print(response.content)
```

```python
# xml 데이터를 dict 형태로  변환
data = xmltodict.parse(response.content)
data
```

```python
## openapi를 이용하여 서버와의 통신하여 데이터를 가져오는 작업

## 어느 서버에서 데이터를 가져와야되는지 주소를 입력
url = "http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson"

## 공공데이터포털에서 주소와 함께 request parameter
## 서비스키
service_key = "/VOjiRaIppLwgf1mXaAkdM6WBLZ/OoWeirK2V1U92jDVjSTAmj16PHQhpVYuumy3EeiuxjYGQxMleNu17eLAyA=="
## 페이지번호
## 페이지당 데이터의 개수
## 시작 시간
## 종료 시간
param = {
    'serviceKey' : parse.unquote(service_key), 
    'pageNo' : '1', 
    'numOfRows' : '5000', 
    'startCreateDt' : '20200101', 
    'endCreateDt' : '20221214'
}


## 여기부터 통신을 시작
response = requests.get(url, params = param)
## response의 데이터의 형태는 xml

## xml데이터를 딕셔너리 형태의 데이터로 변환
result = xmltodict.parse(response.content)

print(result)
```
