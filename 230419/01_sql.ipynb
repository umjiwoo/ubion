### 실습1
1. emp 테이블에서 SAL이 1000이상이고 3000 미만인 사원 정보를 출력
2. emp,dept를 left를 이용해 결합한 뒤 부서의 위치가 "NEW YORK"인 데이터 출력

```python
import pandas as pd
import mod_sql
import imp
```

#### 1. DB 연결해서 쿼리문으로 가져오기

```python
_db = mod_sql.Database("localhost","root","1234","ubion7",3306)
```

```python
sql = """
    select
    *
    from emp
    where (`sal`>=1000) and (`sal`<3000)
"""

_db.sql_query(sql)
```

```python
sql = """
    select
    *
    from emp left join dept
    on emp.deptno=dept.deptno
    where dept.loc="NEW YORK"
"""

_db.sql_query(sql)
```

#### 2. pandas 이용해 가공

```python
df_emp = pd.read_csv("../csv/emp.csv")
df_dept = pd.read_csv("../csv/dept.csv")
```

```python
df_emp.loc[(df_emp["SAL"]>=1000) & (df_emp["SAL"]<3000)].reset_index()
```

```python
df = pd.merge(df_emp,df_dept,how="left",left_on=df_emp["DEPTNO"],right_on=df_dept["DEPTNO"])
df.loc[df["LOC"]=="NEW YORK"]
```

### 실습2
1. emp,dept를 right join으로 결합
2. 부서번호가 40인 데이터만 출력

```python
sql = """
    select
    *
    from emp right join dept
    on emp.deptno=dept.deptno
    where emp.deptno=40 or dept.deptno=40
"""

_db.sql_query(sql)
```
